```python
import sympy as sm
from IPython.display import Markdown as md
C,m,T_e,I,R,kappa,t,T_ex = sm.symbols('C,m,T_e,I,R,\kappa,t,T_{ex}')
T = sm.symbols('T',cls=sm.Function)
```

This is all the symbols we are going to use in the equations. where $ C $ is speci heat capacity, $ T_e $ is the environmental temperature, $ I $ is the current in the metallic strip, $ R $ is the resistance of the metallic strip, $ \kappa $ is the heat transfer coefficient.

This is the temperature function in heating state.

```python
HeatingEq = sm.Eq(C*m*T(t).diff(t),I**2*R-kappa*(T(t)-T_e))
md('$$'+sm.latex(HeatingEq)+'$$'+'\n'+'$$'+sm.latex(sm.dsolve(HeatingEq,T(t),ics={T(0):T_e}))+'$$')
# md('$$'+sm.latex(sm.dsolve(HeatingEq,T(t)))+'$$')
```

```python
CoolingEq = sm.Eq(C*m*T(t).diff(t),-kappa*(T(t)-T_e))
md('$$'+sm.latex(CoolingEq)+'$$'+'\n'+'$$'+sm.latex(sm.dsolve(CoolingEq,T(t),ics={T(0):T_ex}))+'$$')
```
